###### An inner Merge will lose rows that do not match in both dataframes

```python
pd.merge(Fuel_data1, Fuel_data2, how="inner")
```

###### Outer merge returns all rows in both dataframe

```python
pd.merge(Fuel_data1, Fuel_data2, how="outer")
```

###### Remove rows from the right dataframe that do not have a match with the left and keeps all rows from the left

```python
pd.merge(Fuel_data1,Fuel_data2, how='left')
```

###### Concatenation

```python
fuel= pd.concat([Fuel_data,Fuel_data1]).reset_index(drop='True')
```
